```python
data = data.drop(columns=['customerID','gender','MultipleLines','InternetService','Contract', 'PaymentMethod'])
```

```python
data.replace({'Yes': 1, 'No': 0}, inplace=True)
```

```python
data
```

```python
print(data.dtypes)
print('================================================================')
print(data.columns)
```

```python
columns_to_convert = ['OnlineSecurity', 'OnlineBackup', 'DeviceProtection', 'TechSupport', 'StreamingTV', 'StreamingMovies', 'TotalCharges']
data[columns_to_convert] = data[columns_to_convert].apply(pd.to_numeric, errors='coerce')
```

```python
print(data)
print(data.dtypes)
```

```python
data.isnull()
data = data.fillna(0)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report


#declare variables for training
x_data = data.drop(columns=['Churn'])
y_data = data['Churn']

#splitting the dataset
x_train, x_test, y_train, y_test = train_test_split(x_data, y_data, test_size=0.2, random_state=42)

#creating the model
model = LogisticRegression()
model.fit(x_train, y_train)

#evauling the model
y_pred = model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy:.2%}')

print('Classification Report:')
print(classification_report(y_test, y_pred))
```

```python
precision_scores = [0.85, 0.70]
classes = ['No Churn', 'Churn']

# Create a bar plot for precision
plt.bar(classes, precision_scores, color=['red', 'green'])
plt.xlabel('Classes')
plt.ylabel('Precision')
plt.title('Precision for Each Class')
plt.ylim([0, 1])  # Set the y-axis limit to be between 0 and 1
plt.show()
```
